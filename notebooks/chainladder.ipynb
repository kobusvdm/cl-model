```python
import chainladder as cl
import pandas as pd

# Load data
sm = cl.load_sample("clrd")
gb = sm.groupby("LOB")
```

```python
clrd = gb.sum()
X = clrd.loc["medmal", "CumPaidLoss"]
sample_weight = clrd.loc["medmal", "EarnedPremDIR"].latest_diagonal
```

```python

# Specify Model
grid = cl.GridSearch(
    estimator=cl.Pipeline(
        steps=[
            ("dev", cl.Development()),
            ("tail", cl.TailCurve()),
            ("model", cl.Benktander()),
        ]
    ),
    param_grid=dict(
        model__n_iters=list(range(1, 100, 2)), model__apriori=[0.50, 0.75, 1.00]
    ),
    scoring={"IBNR": lambda x: x.named_steps.model.ibnr_.sum()},
    n_jobs=-1,
)
```

```python
type(grid)
```

```python
# Fit Model
grid.fit(X, sample_weight=sample_weight)
```

```python
# Analyze results
output = grid.results_.pivot(
    index="model__n_iters", columns="model__apriori", values="IBNR"
)
print(output.head())
```
